**Business Question 7**

7. Which Transaction Types Have the Highest Failure Rate?

```python
pd.crosstab(
    df['transaction type'],
    df['transaction_status'],
    normalize='index'
) * 100
```

```python
failure_rate = (
    pd.crosstab(
        df['transaction type'],
        df['transaction_status'],
        normalize='index'
    ) * 100
)

failure_rate = failure_rate['FAILED'].sort_values(ascending=False)

plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=failure_rate.index,
    y=failure_rate.values
)

plt.title('Failure Rate by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Failure Rate (%)')

for i, value in enumerate(failure_rate.values):
    ax.text(
        i,
        value,
        f'{value:.2f}%',
        ha='center',
        va='bottom'
    )

plt.show()
```

**Observation**
1. Recharge transactions have the highest failure rate.
2. Bill Payments have the lowest failure rate.
3. All transaction types have success rates close to 95%.

**Business Question 11**

Do Users Spend More on Weekends or Weekdays?

```python
df.groupby('is_weekend')['amount (INR)'].agg(
    ['count','sum','mean']
)
```

```python
weekend_analysis = df.groupby('is_weekend')['amount (INR)'].agg(
    ['count','sum','mean']
)

plt.figure(figsize=(6,4))

ax = sns.barplot(
    x=['Weekday','Weekend'],
    y=weekend_analysis['count']
)

plt.title('Transaction Count: Weekday vs Weekend')
plt.ylabel('Number of Transactions')

for i, value in enumerate(weekend_analysis['count']):
    ax.text(i, value, f'{value:,}', ha='center')

plt.show()
```

```python
plt.figure(figsize=(6,4))

ax = sns.barplot(
    x=['Weekday','Weekend'],
    y=weekend_analysis['sum']
)

plt.title('Revenue: Weekday vs Weekend')
plt.ylabel('Revenue (INR)')

for i, value in enumerate(weekend_analysis['sum']):
    ax.text(i, value, f'₹{value/1e6:.1f}M', ha='center')

plt.show()
```

**Observation**

1. Weekdays account for 178,663 transactions.
2. Weekends account for 71,337 transactions.
3. Weekdays generate significantly more revenue.
4. Average transaction value is almost identical:
- Weekday = ₹1312.52
- Weekend = ₹1309.85

**Business Interpretation**

- Users spend roughly the same amount per transaction regardless of the day.

-->The higher weekday revenue comes mainly because:

- There are more transactions, not because users spend more money.
